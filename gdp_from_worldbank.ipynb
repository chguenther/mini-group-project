```python
# We chose the World Bank API to retrieve a country's GDP for a year.

import requests
import json
import pandas as pd
```

```python

# Ask the user for a country. Need to enter two-letter country code. Return GDP for that country. 
# By default most recent GDP is returned. Ask user what range of years user wants GDP

# Ask User for a country code
print("Welcome to the GDP look up tool.")
country = input("Please enter a two-letter country code.")

# Construct the query url without a year range.
base_url = 'http://api.worldbank.org/v2/country/'
gdp_query = '/indicator/NY.GDP.MKTP.CD?date='
json_request = "&format=json"

url = f"{base_url}{country}{gdp_query}{json_request}"
print(url)

# Retrieve the GDP data from the World bank
response = requests.get(url).json()
#print(json.dumps(response, indent = 3))

# Display response to the user and ask for a year range
# print(json.dumps(response[1], indent=3))
print(f'Most recent year available is {response[1][0]["date"]}.')
print(f'The GDP for that year in billons of dollars is ${response[1][0]["value"]/1000000000:,.2f}')

# Ask user for range of years
print("Retrieve GDP for a number of years.")
start_year = input("What is the start year for which you want the GDP?")
# Check for valid input
end_year = input("What is the end year for which you want to see the GDP?")
# Check for valid input
year_range = str(start_year) + ':' + str(end_year)

# print(year_range)

# Construct new query url for year range.
url = f'{base_url}{country}{gdp_query}{year_range}{json_request}'

# Retrieve response.
response = requests.get(url).json()

try:
    year_list = []
    gdp_list = []
    for year in response[1]:
        year_list.append(year['date'])
        gdp_list.append(year['value']/1000000000)
except:
    print('Data not found.')

data_dict = {'Year': year_list,
             'GDP': gdp_list}

# Convert to a Pandas DataFrame
gdp_df = pd.DataFrame(data_dict)

# Print results
gdp_df

```
